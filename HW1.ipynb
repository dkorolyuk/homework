```python
import pandas as pd
from plotly.graph_objects import *
import numpy as np

books = pd.read_csv('datasets/book-crossing/Books.csv',
                    sep=';',
                    on_bad_lines='skip')
ratings = pd.read_csv('datasets/book-crossing/Ratings.csv',
                      sep=';',
                      on_bad_lines='skip')
```

```python
books.head(10)
```

```python
ratings.head()
```

```python
ratebook = pd.merge(books, ratings, on="ISBN", how="right")
```

```python
ratebook.head()
```

```python
ratebook.drop_duplicates(subset=['Author', 'Title','User-ID'], inplace = True)
```

```python
ratebook.dropna(subset=['ISBN', 'Title'], inplace = True)
```

```python
global_mean = ratebook["Rating"].mean()
```

```python
raw_rate = ratebook.groupby(['Author', 'Title']).mean('Rating')
raw_rate['n'] = ratebook.groupby(['Author', 'Title'])['User-ID'].count()
```

```python
raw_rate
```

```python
raw_rate2 = raw_rate[raw_rate['n']>1]
```

```python
raw_rate2
```

```python
raw_top = raw_rate2.sort_values(by = 'Rating', ascending=False).head(10)
```

```python
raw_top
```

```python
k = 10
raw_rate2['adj_rate'] = (raw_rate2['Rating']*raw_rate2['n'] + global_mean*k)/raw_rate2['n']+k
```

```python
adj_top_tab = raw_rate2.sort_values(by = 'adj_rate', ascending=False).head(10)
```
